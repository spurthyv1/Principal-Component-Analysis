```python
#we are going to construct three models
#1st model is one without principal component analysis
import pandas as pd

df = pd.read_csv("Fundraising.csv")
```

```python
df.describe(include="all")
```

```python
df.info()
```

```python
x = df.iloc[:,0:11]  #column location ..........all rows
y = df['Donor']
```

```python
#preprocess ind variable dataframe x by standardizing it
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
#first model is logistic regression w/o any pca
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x,y, test_size=0.3,
                                                  random_state=0)
```

```python
# Model 1: A simple logistic regressino model without PCA
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train,y_train)
# Evaluating accuracy
model1.score(x_val,y_val)

```

```python
#MODEL 2
#Apply PCA to the entire independent variables(11)......we have 11 PC
#How do we run PC
# Model 2: Logistic regression with PCA
# Model 2: PCA is applied to all independent variables
from sklearn.decomposition import PCA
pca = PCA()
PCA_v1 = pca.fit_transform(x) #first version of PC #we fit model with all ind var
print(PCA_v1[0,:])      #first row
```

```python
#All org ind var now become PC1........11
#i.e 11 org ind var are now represented by 11PC
```

```python
# Check variances explained by each principal components
pca.explained_variance_ratio_
#explained_variance_ratio_ IS proportion of variance explained
#1st PC captures 27%of info
```

```python
#1st PC captured only 26.9% of info in org var........pretty bad
#why so?
#observe data1:08 in video-6
#where do you live? I might be poor living in a wealthy area
#Also not all rich people are generous
########when the data is so diverse............it is very hard to come up with a really good PC that captures all the info
```

```python
# Check eigenvalues of each principal components
pca.explained_variance_
#How many principle components do you want to keep for your predicted model?
#4(threshold=1)
#Instead of using original 11 ind var...with pca we can reduce dimen from 11 to 4
```

```python
# Check eigenvectors of each principal components
pca.components_
#eigenvector tells how principle component is calculated
```

```python
#CREATE SECOND MODEL WITH ONLY TOP 4 PCs
# Only keep PCs with eigenvalue>=1
# Use 4 principal components to train the model
pca = PCA(n_components=4)
PCA_v1 = pca.fit_transform(x)
```

```python
#create model 2
# Data partition: the independent variables are just four PCs
x2_train, x2_val, y_train, y_val = train_test_split(PCA_v1,y, test_size=0.3,
                                                    random_state=0)
```

```python
model2 = LogisticRegression()
model2.fit(x2_train,y_train)
# Evaluating accuracy
model2.score(x2_val,y_val)
```

```python
#after we run pca they become array type of data for numpy..........hce make it a dataframe
#display PCA results
pcaDf = pd.DataFrame(data = PCA_v1
             , columns = ['PC1', 'PC2','PC3','PC4'])
pcaDf.head()
#instead of zipcode and evrything..........we create 4 PCs and these are the numbers that enter the second model
#How do we come up with this value 1.85....?
#Based on Eigenvectors
```

```python
pcaDf.describe()
```

```python
# Two issues with the SECOND MODEL
#1  First PC has lesss variance
#2 Very difficult to explain the results
```

```python
#CREATE THIRD MODEL - apply PCA selectively
# Model 3: Logistic regression with PCA
# Model 3: PCA is applied to selected independent variables(conceptually relevant)
# Convert numpy array to a dataframe
x3 = pd.DataFrame(data = x
             , columns = ['zip', 'homeowner','child','gender','wealth','HV',
                          'icmed','icavg','num','max','totalm'])
x3.corr()
```

```python
# HV,icmed,icavg high corr
#instead of running PCA for everything, apply them selectively
#i.e apply PCA to only correlated variables
```

```python
# Columns without PCA
x3_no_PCA = pd.concat([x3.iloc[:,0:5], x3.iloc[:,8:11]],axis=1)
x3_no_PCA.head()
#np corr between these columns
```

```python
# Columns with PCA
x3_PCA =x3.iloc[:,5:8]
x3_PCA.head()
```

```python
#After selecting cols run PCA
pca2 = PCA()
PCA_v2 = pca2.fit_transform(x3_PCA)
```

```python
# Check variances explained by each principal components
pca2.explained_variance_ratio_
```

```python
#1st PC 88% OF INFO CAPTURED
```

```python
# Check eigenvalues of each principal components
pca2.explained_variance_
```

```python
#WE NEED ONLY ONE AS EV>1
```

```python
# Check eigenectors of each principal components
pca2.components_
```

```python
# PC1=0.54HV+0.59icmed+0.59icavg
#PC1 +VELY CORR WITH HV
```

```python
# Only keep PCs with eigenvalue>=1
# Use PC1 to replace HV, icmed, and icavg
pca2 = PCA(n_components=1)
PCA_v2 = pca2.fit_transform(x3_PCA)
```

```python
# Obtain the PC1
pcaDf = pd.DataFrame(data = PCA_v2
             , columns = ['PC1'])
x3_model = pd.concat([x3_no_PCA,pcaDf],axis=1)
x3_model.head()
```

```python
#pc1=pc for hv,icmed,icavg
#captures 88% of info
```

```python
# Data partition: the independent variables are just four PCs
x3_train, x3_val, y_train, y_val = train_test_split(x3_model,y, test_size=0.3,
                                                    random_state=0)
```

```python
#MODEL 3 CONSTRUCTION
model3 = LogisticRegression()
model3.fit(x3_train,y_train)
# Evaluating accuracy
model3.score(x3_val,y_val)
```

```python
#accuracy dropped a bit but there are advantages
#1 makes more sense
#2 easier to explain the result
```

```python
#Take a look at model estimation
# Show model coefficients
import numpy as np
pd.DataFrame(zip(x3_train.columns, np.transpose(model3.coef_)),
             columns=['features', 'coef'])
```
